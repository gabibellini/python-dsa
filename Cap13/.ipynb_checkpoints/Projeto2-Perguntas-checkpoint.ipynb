## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
#selecionar os dados da categoria office supplies
#calcular a venda total por cidade
#retornar qual é a cidade

df1=df_dsa[(df_dsa.Categoria=='Office Supplies')]
valor_total=df1.groupby('Cidade')['Valor_Venda'].sum()
cidade=valor_total.idxmax() #o método idxmax retorna o maior valor
print('A cidade com maior valor de venda é: ', cidade)
```

### Qual o Total de Vendas por Estado?

```python
#agrupa os dados por estado e calcula o total de vendas + reseta o index e cria reordenado para cada vez
#que eu criar uma fatia do data frame, novo indice criado na sequencia adequada pra não ficar picotado e afetar análise
#matplotlib determina a área do gráfico
#seaborn plota o gráfico de barras de acordo com o dataframe e os eixos x e y
#xticks cria o título no eixo x e a rotação facilita a leitura

df3=df_dsa.groupby('Estado')['Valor_Venda'].sum().reset_index()
plt.figure(figsize=(16,6))
sns.barplot(data= df3, x='Estado', y='Valor_Venda') #ou .set(title='Total de Vendas por Estado')
plt.title('Total de Vendas por Estado')
plt.xticks(rotation=70)
plt.show()
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
#agrupa os valores por segmento e faz a soma dos valores

df5=df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by='Valor_Venda', ascending=False)
df5.head()
```

```python
#o valor foi entregue como notação científica porque o número é muito grande, por isso precisa ajustar o valor
#primeiro precisa converter os dados em valor absoluto - tem como mudar o padrão do pandas ou converter

def autopct_format(values): #função para formatar a saída do pandas
    def my_format(pct):
        total=sum(values)
        val=int(round(pct*total/100.0))
        return ' $ {v:d}'.format(v=val) #ainda inclui o cifrão na saída
    return my_format
```

```python
#plot

#cria a figura
plt.figure(figsize=(16,6))

#gráfico de pizza
#qual a coluna para os valores
#qual os labels
#aqui uso minha função para entregar os valores
plt.pie(df5['Valor_Venda'],
        labels=df5['Segmento'],
        autopct=autopct_format(df5['Valor_Venda']),
        startangle=90)

#limpa o círculo central
centre_circle=plt.Circle((0,0), 0.82, fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)

#labels e anotações
plt.annotate(text='Total de Vendas: '+'$'+str(int(sum(df5['Valor_Venda']))),xy=(-0.25,0))
plt.title('Total de Vendas Por Segmento')
plt.show()   
```
